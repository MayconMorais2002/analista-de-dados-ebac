## 1\. Preços `outliers` de diamante

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1
diamonds = sns.boxplot(x=data["price"])
diamonds.set(title='Preços outliers de diamante', xlabel='Preço');
```

 **No grafico, é possivel observar que apartir dos 12 mil começa a ter outliers.**

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
diamonds = data[["clarity", "carat", "price"]]
diamonds.head()
```

```
with sns.axes_style('whitegrid'):
  #diamonds = diamonds.query("price < 1000")


  grafico = sns.scatterplot(data=diamonds, x="price", y="carat", hue="clarity", palette="pastel")
  grafico.set(title='Correlação entre preço do diamante com seu peso', xlabel='Preço do Diamante', ylabel='Peso do Diamante');
  grafico.get_legend().set_title("Tranparência");
  grafico.figure.set_size_inches(w=30/2.54, h=15/2.54)
```

**Pelo grafico é possivel observar que pela aspecto da tranparência do diamante podemos ver o peso e o seu valor.**

3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```
# gráfico do exercício 3

cars = data["weight"]

with sns.axes_style('whitegrid'):

  grafico = sns.histplot(data=cars, kde=True)
  grafico.set(title='Distribuição contínua aproximada do peso (weight) de carros', xlabel='Peso');
```

**Com o grafico é possivel observar a distribuição do peso dos carros.**
